```python
import matplotlib.pyplot as plt
from cs103 import *
from typing import List

@typecheck
def pie_chart_for_squirrels(filename: str, colour: PrimaryColour) -> None:
    """
    Reads the squirrel data from a given filename, and returns a pie chart showing the percent of squirrels 
    with the given colour seen either at AM or PM.  
    """
    # Template from HtDAP, based on function composition 
    return show_pie_chart(list_of_percents(read(filename), colour)) 
    
@typecheck
def read(filename: str) -> List[SquirrelData]:
    """    
    Reads information from the specified file and returns a list of squirrel data.
    """
    #return []  #stub
    # template based on HtDAP
    # losd contains the result so far
    losd = [] # type: List[SquirrelData]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            if has_colour(row):
                sd = SquirrelData(parse_primary_colour(row[8]), parse_shift(row[4]))
                losd.append(sd)
    
    return losd

@typecheck
def show_pie_chart(fractions: List[float]) -> None:
    """
    Shows a pie chart with the labelled AM or PM and corresponding fractions.
    
    Fractions must each be in the range [0,99] and must sum to 100
    (and so must have at least one element).
    
    Fractions must be a list containing two float values. 
    
    returns None
    """
    #return None  #stub
    # Template based on visualization
    
    plt.pie(fractions, explode=[0, 0.01], labels=["AM","PM"], autopct='%.0f%%', shadow=False, 
            colors = ["#FF9F0F", "#954CC1"], pctdistance = 0.5)

    plt.show()
    
    return None

@typecheck
def list_of_percents(losd: List[SquirrelData], colour: PrimaryColour) -> List[float]:
    """
    Returns a list of percentages of when the total number of the given colour of squirrels were seen
    during the AM (indicated as True) and PM (indicated as False). The AM percent is listed first.
    """
#     return [] # stub
# template from List[SquirrelData]
    # stores the percents so far
    percents = []   # type: List[float]
    
    percents.append(find_percent(losd, colour, True)) # function  find_percent reused from before
    percents.append(find_percent(losd, colour, False))
        
    return percents


start_testing() # tests for main_v2

# should show with about 1/3 labelled AM and about 2/3 labelled PM 
expect(pie_chart_for_squirrels("many_lines.csv", PrimaryColour.GRAY), None)

# should show 50% for both AM and PM for all given colours
expect(pie_chart_for_squirrels("test_all_types_of_squirrels.csv", PrimaryColour.GRAY), None)

# original file, should show both AM and PM with their corresponding percentages
expect(pie_chart_for_squirrels("2018_Central_Park_Squirrel_Census_-_Squirrel_Data-1.csv", PrimaryColour.CINNAMON), None)

summary()

start_testing() # tests for read

expect(read("test_empty.csv"), [])
expect(read("test_one_line_of_data.csv"), [SquirrelData(PrimaryColour.GRAY, False)])
expect(read("test_all_types_of_squirrels.csv"), [SquirrelData(PrimaryColour.CINNAMON, True), 
                                                 SquirrelData(PrimaryColour.GRAY, True),
                                                 SquirrelData(PrimaryColour.BLACK, True),
                                                 SquirrelData(PrimaryColour.CINNAMON, False),
                                                 SquirrelData(PrimaryColour.GRAY, False),
                                                 SquirrelData(PrimaryColour.BLACK, False)])
expect(read("empty_file.csv"), [])

summary()

start_testing() # test for show_pie_chart

# should show a pie chart that is a circle with 78% labelled AM that is orange and 22% labelled PM that is purple. 
expect(show_pie_chart([78, 22]), None)

# should show a pie chart that is a circle with 78% labelled AM that is orange and 22% labelled PM that is purple. 
expect(show_pie_chart([99, 1]), None)

summary()

start_testing() # tests for list_of_percents

expect(list_of_percents([], PrimaryColour.BLACK), [0.0, 0.0])
expect(list_of_percents([], PrimaryColour.GRAY), [0.0, 0.0])
expect(list_of_percents([], PrimaryColour.CINNAMON), [0.0, 0.0])
expect(list_of_percents([SquirrelData(PrimaryColour.GRAY, False), SquirrelData(PrimaryColour.GRAY, True),
                       SquirrelData(PrimaryColour.BLACK, False), SquirrelData(PrimaryColour.CINNAMON, False),
                       SquirrelData(PrimaryColour.GRAY, True), SquirrelData(PrimaryColour.GRAY, False)], 
                       PrimaryColour.GRAY), [1/2*100, 1/2*100])
# 100 % seen in the morning
expect(list_of_percents([SquirrelData(PrimaryColour.BLACK, True), SquirrelData(PrimaryColour.BLACK, True)],
                        PrimaryColour.BLACK), [2/2*100, 0.0])
# 100% seen at night
expect(list_of_percents([SquirrelData(PrimaryColour.CINNAMON, False), SquirrelData(PrimaryColour.CINNAMON, False)],
                       PrimaryColour.CINNAMON), [0.0, 2/2*100])

# random examples

expect(list_of_percents([SquirrelData(PrimaryColour.CINNAMON, True), SquirrelData(PrimaryColour.BLACK, True),
               SquirrelData(PrimaryColour.GRAY, True), SquirrelData(PrimaryColour.CINNAMON, True),
               SquirrelData(PrimaryColour.GRAY, True), SquirrelData(PrimaryColour.BLACK, False),
               SquirrelData(PrimaryColour.CINNAMON, False), SquirrelData(PrimaryColour.BLACK, False),
               SquirrelData(PrimaryColour.GRAY, True), SquirrelData(PrimaryColour.CINNAMON, True),
               SquirrelData(PrimaryColour.BLACK, True), SquirrelData(PrimaryColour.CINNAMON, False)],
              PrimaryColour.CINNAMON), [60.0, 40.0])
expect(list_of_percents([SquirrelData(PrimaryColour.GRAY, False), SquirrelData(PrimaryColour.GRAY, False),
               SquirrelData(PrimaryColour.GRAY, False), SquirrelData(PrimaryColour.CINNAMON, True),
               SquirrelData(PrimaryColour.BLACK, True), SquirrelData(PrimaryColour.CINNAMON, False),
               SquirrelData(PrimaryColour.BLACK, False), SquirrelData(PrimaryColour.GRAY, True),
               SquirrelData(PrimaryColour.GRAY, True)], PrimaryColour.GRAY), [40.0, 60.0])

expect(list_of_percents([SquirrelData(PrimaryColour.CINNAMON, True), SquirrelData(PrimaryColour.CINNAMON, False), 
                        SquirrelData(PrimaryColour.BLACK, True), SquirrelData(PrimaryColour.GRAY, False), 
                        SquirrelData(PrimaryColour.CINNAMON, True), SquirrelData(PrimaryColour.BLACK, False),
                        SquirrelData(PrimaryColour.CINNAMON, False), SquirrelData(PrimaryColour.CINNAMON, False),
                        SquirrelData(PrimaryColour.GRAY, True), SquirrelData(PrimaryColour.GRAY, False), 
                        SquirrelData(PrimaryColour.CINNAMON, False), SquirrelData(PrimaryColour.CINNAMON, False)], 
                       PrimaryColour.CINNAMON), [2/7*100, 5/7*100])

summary()
```
